```python
# Loading necessary packages
import pandas as pd
```

```python
# File paths. Change according to your local directory paths
code_repo_dir = '/Users/user/Documents/Codes/H5N1/H5N1-State-and-County-Simulations-and-Analysis/'

input_sub_dir = 'data/raw/'
input_filename = 'cattle_FLAPS_for_USAMMv3.txt'
input_filepath = code_repo_dir + input_sub_dir + input_filename

output_sub_dir = 'data/processed/'
output_filename = 'dairy_cows_by_premises.csv'
output_filepath = code_repo_dir + output_sub_dir + output_filename
```

```python
# Load the FLAPS file that contain information of cattle per premises (eg. farm)
cattle_by_premises = pd.read_csv(input_filepath, sep = "\t")
cattle_by_premises[cattle_by_premises.type == "d"]
```

```python
# Filter rows by cattle type 'd' (for dairy cow) and select the columns: Premises Id (Id), 
# County FIPS (County), Lat, Lon, Number of dairy cattle in premises (d) 
dairy_cows_by_premises = cattle_by_premises.loc[
    cattle_by_premises.type == "d",  ['Id', 'County', 'Lat', 'Lon', 'd']
].reset_index(drop=True)
dairy_cows_by_premises.to_csv(output_filepath, index=False)
```
