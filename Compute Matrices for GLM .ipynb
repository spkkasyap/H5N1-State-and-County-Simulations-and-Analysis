```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap as Basemap
import geopandas as gpd
import numpy as np
import os
import glob
import re
```

```python
valid_states = [
    'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA',
    'MA', 'MD', 'ME', 'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY',
    'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY'
]
print(len(valid_states))
```

```python
directory_path = "data/USAMMv3_cattle_networks/dairy"

data = []
count = 0
epsilon = 0.0001
unique_states = set()

for file_path in glob.glob(os.path.join(directory_path, "dairy_network_*.network")):
    filename = os.path.basename(file_path)  
    match = re.search(r"dairy_network_(\d+)\.network", filename)
    if match:
        file_number = int(match.group(1))  
    count += 1

    if count %50 ==0:
        print(f'{count} files processed.')
    print(file_path)
    df_2009 = pd.read_csv(file_path, sep = "\t")
    df_2009.rename(columns = {"oStateAbbr": "org_st", "dStateAbbr": "dest_st"}, inplace=True)
    df_2009['nb_shp'] = 1
    df_2009 = df_2009.groupby(['org_st', 'dest_st'])['nb_shp'].sum().reset_index()
    #df_2009 = df_2009.loc[df_2009.org_st != df_2009.dest_st]
    
    full_state_pairs = pd.MultiIndex.from_product([valid_states, valid_states], names=['org_st', 'dest_st']).to_frame(index=False)
    df_2009 = full_state_pairs.merge(df_2009, on=['org_st', 'dest_st'], how='left')
    df_2009['nb_shp'] = df_2009['nb_shp'].fillna(0)

    df_2009['nb_shp_norm_2009'] = ((df_2009['nb_shp'] - df_2009['nb_shp'].min()) / (df_2009['nb_shp'].max() - df_2009['nb_shp'].min())) * (1-epsilon) + epsilon
    #print(df_2009.head(10))

    matrix_df = df_2009.pivot_table(index="org_st", columns="dest_st", values="nb_shp_norm_2009", fill_value=0)

    
    matrix_df.to_csv(f'matrices2009/year/state_shipping_matrix_{file_number}.csv')
    #print(matrix_df.head())  
    print(matrix_df.shape)
```

```python
directory_path = "data/USAMMv3_cattle_networks/dairy"

data = []
count = 0
epsilon = 0.0001
month = 12

start_days = {1: 1, 2: 32, 3: 60, 4: 91, 5: 121, 6: 152, 7: 182, 8: 213, 9: 244, 10: 274, 11: 305, 12: 335}

end_days = {1: 31, 2: 59, 3: 90, 4: 120, 5: 151, 6: 181, 7: 212, 8: 243, 9: 273, 10: 304, 11: 334, 12: 365}

for file_path in glob.glob(os.path.join(directory_path, "dairy_network_*.network")):
    filename = os.path.basename(file_path)  
    match = re.search(r"dairy_network_(\d+)\.network", filename)
    if match:
        file_number = int(match.group(1))  
    count += 1
    if count %50 ==0:
        print(f'{count} files processed.')
    print(file_path)
    s_day = start_days[month]
    e_day = end_days[month]
    df_2009 = pd.read_csv(file_path, sep = "\t")
    df_2009.rename(columns = {"oStateAbbr": "org_st", "dStateAbbr": "dest_st"}, inplace=True)
    df_2009['nb_shp'] = 1
    df_2009 = df_2009.loc[(df_2009['dayOfYear'] >= s_day) & (df_2009['dayOfYear'] <= e_day)]
    df_2009 = df_2009.groupby(['org_st', 'dest_st'])['nb_shp'].sum().reset_index()
    
    full_state_pairs = pd.MultiIndex.from_product([valid_states, valid_states], names=['org_st', 'dest_st']).to_frame(index=False)
    df_2009 = full_state_pairs.merge(df_2009, on=['org_st', 'dest_st'], how='left')
    df_2009['nb_shp'] = df_2009['nb_shp'].fillna(0)
    
    df_2009['nb_shp_norm_2009'] = ((df_2009['nb_shp'] - df_2009['nb_shp'].min()) / (df_2009['nb_shp'].max() - df_2009['nb_shp'].min())) * (1-epsilon) + epsilon    
    matrix_df = df_2009.pivot_table(index="org_st", columns="dest_st", values="nb_shp_norm_2009", fill_value=0)
    matrix_df.to_csv(f'matrices2009/month{month}/state_shipping_matrix_{file_number}_month_{month}.csv')    
    #print(matrix_df.head())  
    print(matrix_df.shape)
```
